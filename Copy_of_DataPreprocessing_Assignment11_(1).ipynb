**Step 1: Importing the libraries**

```python
import pandas as pd
```

**Step 2: Importing dataset**

```python
df=pd.read_csv("/content/drive/MyDrive/Task7/Assignment11_to_16/Data.csv")
```

```python
df
```

**Step 3: Handling the missing data**

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.Age=df.Age.fillna(int(df.Age.mean()))
```

```python
df.Salary=df.Salary.fillna(int(df.Salary.median()))
```

```python
df
```

**Step 4: Encoding categorical data**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df1=df
df1
```

```python
df1.Purchased=encoder.fit_transform(df1.Purchased)
```

```python
df1
```

**Step 5: Creating a dummy variable**

```python
dum=pd.get_dummies(df1.Country)
```

```python
df2=pd.concat([df1,dum],axis="columns")
df3=df2.drop(columns=["Country"])
df3
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
x=df3.drop("Purchased",axis="columns")
y=df3.Purchased
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
xtrain=scaler.fit_transform(x_train)
xtest=scaler.transform(x_test)
```
